```python
#Imports
import requests
import pandas as pd
import datetime
def get_initial_value(lastvalue,percentChange):
    print(percentChange,lastvalue)
    print((float(percentChange) * float(lastvalue)))
    return ((float(percentChange) * float(lastvalue)) + float(lastvalue))
def job():
    print('fetchin data')
    response = requests.get("https://poloniex.com/public?command=returnTicker").json()
    for coin_name, value in response.items():
        if coin_name.endswith('_BTC'):
            data.setdefault('Moeda', []).append(coin_name.replace('_BTC',''))
            data.setdefault('Periodicidade', []).append('1 min')
            data.setdefault('Datetime', []).append(datetime.datetime.now())
            data.setdefault('Open', []).append(get_initial_value(value['last'],value['percentChange']))
            data.setdefault('Low', []).append(value['lowestAsk'])
            data.setdefault('High', []).append(value['highestBid'])
            data.setdefault('Close', []).append(value['last'])
```

```python
import schedule
import time
data = {}
    
schedule.every(1).minutes.do(job)

while 1:
    schedule.run_pending()
    time.sleep(1)
```
